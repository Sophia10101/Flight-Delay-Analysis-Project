```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_delays[["carrier", "carrier_name"]].drop_duplicates().sort_values(by="carrier")
```

```python
#Plot for delays per airline 
plt.figure(figsize=(12,6))
plt.title("Delays per Carrier in December 2024 and January 2025")
plt.ylabel("Number of delays")
sns.barplot(df_delays, x="carrier", y="arr_del15",hue="carrier")
```

This plot shows that Southwest Airlines had the most delays with American Airlines as a runner up and Mesa Airlines had the least number of delays.

```python
#Flights per airline 
plt.figure(figsize=(12,6))
plt.title("Flights per Carrier in December 2024 and January 2025")
plt.ylabel("Number of flights")
sns.barplot(df_delays, x="carrier", y="arr_flights",hue="carrier")
```

This plot shows that Southwest Airlines had the most flights by a majority. American airlines had the second most. This is interesting because it shows a correlation between number of flights and delays. From the two graphs it seems that larger or smaller number of flights impacts delays. Number of flights needs to be taken into consideration with number of delays for the probability of a delay occuring. 

```python
#Plot for delays per airline 
plt.figure(figsize=(12,6))
plt.title("Cancelled flights per Carrier in December 2024 and January 2025")
plt.ylabel("Number of cancelled flights")
sns.barplot(df_delays, x="carrier", y="arr_cancelled",hue="carrier")
```

This shows that Southwest and American airlines were most likely to cancel a flight.

```python
# Delays per month
#Flights per airline 
plt.figure(figsize=(12,6))
plt.title("Flights delayed per month")
plt.ylabel("Number of delays")
sns.barplot(df_delays, x="month", y="arr_flights",hue="month")
```

This plot shows that July had the most delays in the 2024-2025 year. 

```python
#Overall contribution of each cause to a delay 
values = df_delays[['carrier_delay', 'weather_delay', 'nas_delay', 'security_delay', 'late_aircraft_delay']].sum()
labels = ["Airlines fault", "Weather Delay", "NAS delay", "Security Delay", "Late Aircraft delay"]
plt.figure(figsize=(8,8))
plt.pie(values,labels=labels, autopct='%1.1f%%',pctdistance=0.75)
```
